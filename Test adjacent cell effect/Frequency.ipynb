```python
import meep as mp
import numpy as np
import matplotlib.pyplot as plt

# 定義模擬參數
resolution = 50  # pixels/μm，模擬空間中的解析度
dpml = 1.0  # PML (Perfectly Matched Layer) 厚度，用於吸收邊界
dsub = 0.5  # 衬底 (substrate) 厚度
dca = 0.06  # 空氣區域 (air region) 在兩個單元格 (grating cell) 之間的厚度
gp = 0.2  # 單元格週期性 (grating periodicity)
gh = 0.6  # 光柵高度 (grating height)
gw = 0.08  # 光柵單元格 (grating cell) 寬度
num_cells = 3  # 光柵的單元格數量
dair = 14  # 光柵旁邊的空氣區域厚度
wvl = 0.532  # 中心波長 (center wavelength)
fcen = 1 / wvl  # 中心頻率 (center frequency)，假設 meep 的 C=1，所以 1/λ = f，?即是 2
glass = mp.Medium(index=2.6678)  # 光柵材料，把 paper 裡的材料在對應波長下拿去查詢找到的折射率值
pml_layers = [mp.PML(thickness=dpml, direction=mp.X)]  # PML 設定，方向為 X
symmetries = [mp.Mirror(mp.Y)]  # Y 軸對稱

sx = dpml + dsub + gh + dair + dpml  # 模擬空間 x 方向的大小
sy = gp * num_cells  # 模擬空間 y 方向的大小
cell_size = mp.Vector3(sx, sy)  # 模擬空間大小

sources = [
    mp.Source(
        mp.ContinuousSource(fcen),  # 頻率為 fcen 的連續光源
        component=mp.Ez,  # 光場的分量為 Ez
        center=mp.Vector3(-0.5 * sx + dpml),  # 光源的中心位置，x 方向偏移 -0.5 * sx + dpml
        size=mp.Vector3(y=sy),  # 光源的大小，只有 y 方向有大小，大小為 sy
    )
]

geometry = [
    mp.Block(
        material=glass,
        size=mp.Vector3(dsub, mp.inf, mp.inf),
        center=mp.Vector3(-0.5 * sx + dpml + dsub / 2),
    )
]

sim = mp.Simulation(
    cell_size=cell_size,
    boundary_layers=pml_layers,
    geometry=geometry,
    sources=sources,
    resolution=resolution,
    symmetries=symmetries,
)

for j in range(num_cells):
    geometry.append(
        mp.Block(
            material = glass,
            size = mp.Vector3(gh, gw, mp.inf),
            center = mp.Vector3(
                -0.5 * sx + dpml + dsub + 0.5 * gh,
                -0.5 * sy + gp/2 + (j * gp),
            ),
        )
    )
```

```python
%matplotlib inline
plt.figure(dpi = 100)
sim.plot2D()
plt.show()
```

```python
# 執行模擬
sim.k_point = mp.Vector3()
sim.run(until=100)

sim.run(until = 100)
plt.figure(dpi = 100)
sim.plot2D(fields=mp.Ez)
plt.show()
```

```python

# 取得 Ez 場量在一段時間內的資料
ez_data = np.array(sim.get_array(component=mp.Ez, cmplx = True))

# 計算時間步長及頻率間隔
dt = 0.0001
df = 1.0 / (len(ez_data) * dt)

# 進行 FFT
ez_fft = np.fft.fft(ez_data)
ez_fft_magnitude = np.abs(ez_fft)
frequencies = np.fft.fftfreq(len(ez_data), dt)

# 繪製傅立葉轉換結果
plt.figure(dpi=100)
plt.plot(frequencies, ez_fft_magnitude)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.show()

print(ez_data)
```
